```
import functools

def answer(a):
  neg = list(filter(lambda x: x<0, a))
  pos = list(filter(lambda x: x>0, a))
  if (len(neg) == 1 and len(pos) == 0 and 0 in a) or (len(neg) == 0 and len(pos) == 0 and 0 in a):
    return 0  
  if len(neg) == 0:
    neg_prod = 1
  elif len(neg) == 1 and len(pos) == 0 and 0 not in a:
    neg_prod = neg[0]
    return neg_prod
  elif (len(neg) == 1 and len(pos) == 0 and 0 in a) or (len(neg) == 0 and len(pos) == 0 and 0 in a):
    return 0
  else:
    neg_prod = functools.reduce(lambda x,y : x*y ,neg)
  if len(pos) == 0:
    pos_prod = 1
  else:
    pos_prod = functools.reduce(lambda x,y : x*y ,pos)
  if len(neg) % 2 == 0:
    return str(neg_prod * pos_prod )
  else:
    return str(neg_prod * pos_prod // max(neg))
print(answer([-2,-1,0,-3]))
```

```
import math
import os
import random
import re
import sys

#
# Complete the 'surfaceArea' function below.
#
# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY A as parameter.


def is_valid(a,b,rows,columns):
    return (a >= 0) and (b >= 0) and (a < rows) and (b < columns)

def surfaceArea(A):
    rows = len(A)
    columns = len(A[0])
    area = 2*rows*columns

    for i in range(rows):
        for j in range(columns):
            if is_valid(i-1,j,rows,columns):
                if A[i-1][j] < A[i][j]:
                    area += A[i][j] - A[i-1][j]
            else:
                area += A[i][j]

            if is_valid(i,j-1,rows,columns):
                if A[i][j-1] < A[i][j]:
                    area += A[i][j] - A[i][j-1]
            else:
                area += A[i][j]

            if is_valid(i,j+1,rows,columns):
                if A[i][j+1] < A[i][j]:
                    area += A[i][j] - A[i][j+1]
            else:
                area += A[i][j]

            if is_valid(i+1,j,rows,columns):
                if A[i+1][j] < A[i][j]:
                    area += A[i][j] - A[i+1][j]
            else:
                area += A[i][j]                           

    return area
A = [[1,3,4],[2,2,3],[1,2,4]]
print(surfaceArea(A))
```

```
# Python3 program to find all combinations that
# sum to a given value
sum = int(input())

num_denom = int(input())
arr = []
for i in range(num_denom):
  arr.append(int(input()))

def combinationSum(arr, sum):
	ans = []
	temp = []

	# first do hashing nothing but set{}
	# since set does not always sort
	# removing the duplicates using Set and
	# Sorting the List
	arr = sorted(list(set(arr)))
	findNumbers(ans, arr, temp, sum, 0)
	return ans

def findNumbers(ans, arr, temp, sum, index):
	
	if(sum == 0):
		
		# Adding deep copy of list to ans
		ans.append(list(temp))
		return
	
	# Iterate from index to len(arr) - 1
	for i in range(index, len(arr)):

		# checking that sum does not become negative
		if(sum - arr[i]) >= 0:

			# adding element which can contribute to
			# sum
			temp.append(arr[i])
			findNumbers(ans, arr, temp, sum-arr[i], i)

			# removing element from list (backtracking)
			temp.remove(arr[i])



ans = combinationSum(arr, sum)
print(len(ans))
```

```
def combinationSum(self, candidates: List[int], target: int):
        ans = set()
        def bt(choices, index, nums, rem):
            nonlocal ans
            if rem == 0:
                ans.add(nums)
                return
            if rem < 0:
                return
            if index >= len(choices):
                return
            else:
                bt(choices, index, nums+(choices[index],), rem-choices[index])
                bt(choices, index+1, nums, rem)
            
        bt(candidates, 0, (), target)
        return ans
```

```
def matrixRotation(matrix, r):

    while r>0:
        rows = len(matrix)
        columns = len(matrix[0]) 
        dim = [rows,columns]
        m,n = 0,0
        while m < min(dim)//2:
            start = matrix[m][n]
            start_index = [m,n]
            i = m
            j = n
            
            while j+1 < columns:
                if [i,j+1] == start_index:
                    matrix[i][j] = start
                else:
                    matrix[i][j] = matrix[i][j+1]
                j += 1

            while i+1 < rows:
                if [i+1,j] == start_index:
                    matrix[i][j] = start
                else:
                    matrix[i][j] = matrix[i+1][j]
                i += 1

            while j-1 >= n:
                if [i,j-1] == start_index:
                    matrix[i][j] = start
                else:
                    matrix[i][j] = matrix[i][j-1]
                j -= 1

            while i-1 >= m:
                if [i-1,j] == start_index:
                    matrix[i][j] = start
                else:
                    matrix[i][j] = matrix[i-1][j]
                i -= 1
            m += 1
            n += 1
            rows -= 1
            columns -= 1
        r -= 1
    for i in matrix:
        for j in i:
            print(j,end = " ")
        print()
```

```
def numDistinct(self, s: str, w: str) -> int:
    d={-1:1}
    n = len(s)
    m = len(w)

    if n < m:
        return 0
    for i in range(n):
        for j in reversed(range(m)):
            if w[j] == s[i]:
                d[j] = d.get(j,0)+d.get(j-1, 0)

    return d.get(m-1,0)
```
